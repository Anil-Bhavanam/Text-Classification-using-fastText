```python
import pandas as pd
data=pd.read_csv("/content/ecommerce_dataset.csv",names=['category','description'],header=None)
data.head()
```

```python
data.category.value_counts()
```

```python
data.dropna(inplace=True)
```

```python
data.category.replace("Clothing & Accessories","Clothing_Accessories",inplace=True)
data.category.unique()
```

```python
data['category']='__label__'+data['category'].astype(str)
data.head()
```

```python
data['category_description']=data['category']+' '+data['description']
data=data.drop('description',axis=1)
data.head()
```

# removing the punctuation and anything not a word will replace with space

```python
import re

def preprocess(text):
  text=re.sub(r"[^\w\s\']"," ",text)
  text=re.sub(r"[ \n]+"," ",text)
  return text.strip().lower()
```

```python
data['category_description']=data['category_description'].map(preprocess)
data.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data.category_description,train_size=0.8)
```

```python
train.to_csv("train_csv",header=False,index=False)
test.to_csv("test_csv",header=False,index=False)
```

```python
import fasttext
model=fasttext.train_supervised('train_csv')
```

```python
model.test('test_csv')
```
